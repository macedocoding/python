# Linguagem de marcação Markdown

## Sintaxe básica de Markdown 

Hierarquia de títulos é relacionada com a quantidade de # utilizada. Pode-se utilizar até 6.

Para compatibilidade, pular uma linha antes e depois dos títulos.

Não há marcação para parágrafos, apenas comece a digitar, porém sem indentação inicial.  
Mudanças de linha não configuram, necessariamente, quebra de linha. Para isso, é necessário inserir 2 ou mais espaços ao final de uma linha OU utilizar a tag HTML br.

Inserir uma ou mais linhas em branco entre blocos de texto, pula somente uma linha.

Para deixar um trecho em negrito, envolva-o com ** ou __ no início e no fim (sem espaços). **Faça dessa forma** ou __dessa forma__, porém dê preferência aos asteriscos por questão de compatibilidade.

Para deixar um trecho em itálico, envolva-o com * ou _ no início e no fim (sem espaços). *Faça dessa forma* ou _dessa forma_, porém dê preferência aos asteriscos por questão de compatibilidade.

Para deixar um trecho em negrito E itálico, envolva-o com ** ou __ no início e no fim (sem espaços). ***Faça dessa forma*** ou ___dessa forma___, porém dê preferência aos asteriscos por questão de compatibilidade.

> Para a criação de uma citação longa (blockquote), deve-se inserir o símbolo > no início de um bloco de texto, preferencialmente com uma linha em branco antes e depois do trecho.

> Pode-se adicionar múltiplos parágrafos em um único blockquote inserido-se > na linha em branco logo abaixo o primeiro bloco de texto.
>
> O resultado final é esteticamente agradável.

> Pode-se criar blockquotes aninhados adicionando um > adicional para cada nível.
>
>> Este bloco de texto está aninhado em relação ao anterior (utilizando >>).
>
>>> E assim sucessivamente (utilizando >>>).

> ### Utilização de elementos dentro do blockquote
>
> É possível a utlização de outros elementos dentro do blockquote (não todos), como o título de terceira hierarquia acima ou ***este texto em negrito e itálico.***
>
> - Listas, inclusive;
> - Outro item;
> - Outro item.

Listas ordenadas começam com o número 1 seguido de um ponto (ou fechamento de parânteses, porém utilize ponto por questão de compatibilidade) e um espaço. Os números subsequentes não precisam estar em ordem, apenas o primeiro item deve possuir o índice 1.  
Listas aninhadas são indentadas um nível em relação à sua lista original e também devem iniciar com 1.

1. Primeiro item da lista;
2. Segundo item;
    1. Primeiro item de uma sublista;
        1. Primeiro item de uma sub sublista;
    2. Segundo item de uma sublista.
3. Terceiro.

Para criar listas não ordenadas, adicione - ou + ou * seguido de um espaço. Utilize apenas um tipo de símbolo em uma lista para manter a coerência e impedir linhas em branco indesejáveis.  
Para criar listas aninhadas, indente um ou mais itens.

- Coisa
- Outra coisa
- Mais uma
    - Item
    - Item 2
        - Oi
- Pra fechar

Para adicionar elementos entre os itens de uma lista sem perturbar sua continuidade, adicione o elemento com uma indentação (ou 4 espaços) entre os elementos da lista, além de uma linha em branco antes e depois.

* Linux é utilizado por empresas do mundo todo;
* O mascote adotado pela comunidade Linux é o Tux;

    ![Tux, o mascote Linux](https://d33wubrfki0l68.cloudfront.net/e7ed9fe4bafe46e275c807d63591f85f9ab246ba/e2d28/assets/images/tux.png)

* Uma infinidade de distribuições está disponível para download.

Se quiser denotar código com sua linha de texto, coloque-a entre crases e ela será exibida com fonte monoespaçada propícia para códigos.

`print("Olá, mundo!")`

Caso queira imprimir na tela o símbolo crase, utilize duas crases para delimitar seu código.

``console.log(`Meu nome é ${nome}.`);``

Para inserir blocos inteiros de código, apenas indente (ou adicione 4 espaços) o bloco inteiro pelo menos uma vez.

    #include <stdio.h>

    int main() {
        printf("Olá, mundo!");
        return 0;
    }

Para adicionar uma linha horizontal, coloque, no mínimo, ***, --- ou ___.  
Não se esqueça de adicionar uma linha em branco entes e depois, caso contrário, se utilizar a notação ---, o texto imediatamente antes será interpretado como um título de segunda ordem.

***
